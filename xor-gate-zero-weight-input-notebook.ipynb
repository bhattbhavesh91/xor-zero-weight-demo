```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")
```

# Xor Model with Zeros

```python
zero_model = Sequential()
zero_model.add(Dense(16, 
                input_dim=2, 
                activation='relu',
                kernel_initializer='zeros'))
zero_model.add(Dense(1, activation='sigmoid'))
```

```python
zero_model.compile(loss='mean_squared_error',
                       optimizer='adam',
                       metrics=['binary_accuracy'])
zero_model.fit(training_data, target_data, epochs=1000)
```

```python
scores = zero_model.evaluate(training_data, target_data)
print("\n%s: %.2f%%" % (zero_model.metrics_names[1], scores[1]*100))
print (zero_model.predict(training_data).round())
```

# Model With Random Weights

```python
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error',
                       optimizer='adam',
                       metrics=['binary_accuracy'])
model.fit(training_data, target_data, epochs=1000)
scores = model.evaluate(training_data, target_data)

print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print (model.predict(training_data).round())
```
